```python
import json
import pandas as pd

data_path = "MathV.jsonl"
question_dataset = pd.read_json(data_path, lines=True)
```

```python
question_dataset_mini = pd.read_json("MathVMini.jsonl", lines=True)
```

```python
difficulty = question_dataset["level"]
difficulty_mini = question_dataset_mini["level"]
```

```python
import numpy as np
print(np.mean(difficulty))
print(np.mean(difficulty_mini))
```

```python
import matplotlib.pyplot as plt

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot distribution of difficulty
difficulty.value_counts().sort_index().plot(kind='bar', ax=axes[0], color='steelblue')
axes[0].set_title('Distribution of Difficulty (MathV)')
axes[0].set_xlabel('Difficulty Level')
axes[0].set_ylabel('Count')
axes[0].grid(axis='y', alpha=0.3)

# Plot distribution of difficulty_mini
difficulty_mini.value_counts().sort_index().plot(kind='bar', ax=axes[1], color='coral')
axes[1].set_title('Distribution of Difficulty (MathVMini)')
axes[1].set_xlabel('Difficulty Level')
axes[1].set_ylabel('Count')
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```
